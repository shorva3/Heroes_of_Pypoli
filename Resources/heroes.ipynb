```python
import pandas as pd
import numpy as np

file = "purchase_data.csv"
df = pd.read_csv(file, encoding = "utf8")
df.head()
```

```python
df[['SN']].describe()
```

```python
p_c = df['SN'].nunique()
```

```python
player_count = pd.DataFrame([{"Total Players":df['SN'].nunique()}])
player_count
```

```python
uitems = df['Item ID'].nunique()
app = round(df['Price'].mean(),2)
total_p = df['Price'].count()
tot_rev = df['Price'].sum()


purchase_analysis = pd.DataFrame({"Number of Unique Items":uitems,
                                 "Average Purchase Price":app,
                                 "Total Number of Purchases":total_p,
                                 "Total Revenue":tot_rev}, index = [0])

purchase_analysis.head()
```

### Gender Demographics

```python
total_count = df['SN'].count()
gender_count = df.groupby('Gender') ['SN'].nunique()
gender_percent = round((gender_count / p_c) *100,2)

gender_demo = pd.DataFrame({"Percentage of Players": gender_percent, "Total Count":gender_count})

gender_demo
```

## Gender Purchase Analysis

```python
genderdf = df.groupby(['Gender'])
purchase_count = genderdf['Purchase ID'].count()
avg_purchase = round(genderdf['Price'].mean(),2)
total_purchase_value = genderdf['Price'].sum()
per_person = round(total_purchase_value / genderdf['SN'].nunique(),2) 

```

```python
gender_purchase_analysis = pd.DataFrame({"Purchase Count":purchase_count, 
                                         "Average Purchase Price":avg_purchase,
                                        "Total Purchase Value":total_purchase_value,
                                        "Average Purchase Total":per_person})

gender_purchase_analysis
```

## Age Demo

```python
## Create new Data Frame with Age Bracket bins
bins = [0,9,14,19,24,29,34,39,40]
value_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
df['Age Bracket'] = pd.cut(df['Age'],bins, labels = value_labels)
```

```python
df_age = df.groupby(['Age Bracket'])
## Find purchase count
pc = df_age['Price'].count()
##Average Purhcase Price
avg_pp = round(df_age['Price'].mean(),2)
## Total purchase value
total_pp = df_age['Price'].sum()
##avg purchase total
true_pt = round(total_pp / df_age['SN'].nunique(),2)
```

## Age Demographic Analysis

```python
age_demo_analysis = pd.DataFrame({"Purchase Count":pc,
                                 "Average Purchase Price": avg_pp,
                                 "Total Purchase Value": total_pp,
                                 "Average Purchase Total per Person by Age Group": true_pt})
age_demo_analysis
```

## Top Spender Analysis

```python
top_spend = df.groupby(['SN'])
##group by screename
total_p = top_spend['Price'].sum()
##sum up spenders
total_c = top_spend['Price'].count()
##count up top spenders
total_avg = top_spend['Price'].mean()
##avg of top spenders

top_spend_df = pd.DataFrame({"Purchase Count":total_c,
                            "Average Purchase Price":round(total_avg,2),
                            "Total Purchase Value":total_p})
top_spend_df_sort = top_spend_df.sort_values('Total Purchase Value', ascending = False)
##insert into dataframe and sort ascending
top_spend_df_sort.head(5)
```

## Most Popular Item Analysis

```python
##Group by Item ID and Name
pop_items = df.groupby(['Item ID','Item Name'])
## Items count
pop_count = pop_items['Item ID'].count()
##Sum of pop items
pop_sum = pop_items['Price'].sum()
##Find mean of items which is just price
pop_avg = pop_items['Price'].mean()
##insert into dataframe
most_popular = pd.DataFrame({"Purchase Count":pop_count,
                            "Item Price":pop_avg,
                            "Total Purchase Value":pop_sum})
##sort by ascending
most_popular_sort = most_popular.sort_values('Purchase Count', ascending = False)
most_popular_sort.head()
```

## Most Profitable Items

```python
most_popular_profit = most_popular.sort_values('Total Purchase Value', ascending = False)
most_popular_profit.head()
```
